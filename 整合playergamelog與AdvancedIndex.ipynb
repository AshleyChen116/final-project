# 整合球員資料:一般數據加上進階數據(player_game + player_advanced_stats)

```python
import pandas as pd
```

# 讀取球員比賽數據

```python
playergame = pd.read_csv("D:/Player_game.csv")
```

# 整理欄位

```python
playergame = playergame.rename(columns={"Player_ID": "PLAYER_ID", "Game_ID": "GAME_ID"})
player_game = playergame[['PLAYER_ID', 'GAME_ID',  'WL',
       'MIN', 'FG_PCT', 'FG3_PCT',
       'FT_PCT', 'OREB', 'DREB', 'AST', 'STL', 'BLK', 'TOV', 'PF',
       'PTS', 'PLUS_MINUS']]
player_game

playergame = player_game.dropna()
```

# 讀取進階數據

```python
advanced = pd.read_csv("D:/advanced_stats (1).csv")
advanced
```

# 整理重複的row與欄位

```python
advanced.drop_duplicates(keep = 'first', inplace = True) 
advanced = advanced[['PLAYER_ID', 'GAME_ID', 'TEAM_ID',
       'OFF_RATING', 'DEF_RATING', 
       'NET_RATING', 'AST_PCT', 'AST_TOV', 'AST_RATIO', 'OREB_PCT', 'DREB_PCT',
       'REB_PCT', 'TM_TOV_PCT', 'EFG_PCT', 'TS_PCT', 'USG_PCT',
        'PACE', 'POSS', 'PIE']]
```

# 合併兩個dataframe

```python
player_game = player_game.merge(advanced,  on=['PLAYER_ID', 'GAME_ID'], how='inner')
```

```python
player_game.drop_duplicates(keep = 'first', inplace = True) 
player_game
```

# 找出每個球季的數據，每季做成一個dataframe

```python
# 2013~2018
player_game_adv_2013 = player_game[player_game["GAME_ID"].apply(lambda x: str(x).startswith('213'))]
player_game_adv_2013.to_csv(r'C:\Users\ACER\aiwork\final\train\player_game_adv_2013.csv', index = False)

player_game_adv_2014 = player_game[player_game["GAME_ID"].apply(lambda x: str(x).startswith('214'))]
player_game_adv_2014.to_csv(r'C:\Users\ACER\aiwork\final\train\player_game_adv_2014.csv', index = False)

player_game_adv_2015 = player_game[player_game["GAME_ID"].apply(lambda x: str(x).startswith('215'))]
player_game_adv_2015.to_csv(r'C:\Users\ACER\aiwork\final\train\player_game_adv_2015.csv', index = False)

player_game_adv_2016 = player_game[player_game["GAME_ID"].apply(lambda x: str(x).startswith('216'))]
player_game_adv_2016.to_csv(r'C:\Users\ACER\aiwork\final\train\player_game_adv_2016.csv', index = False)

player_game_adv_2017 = player_game[player_game["GAME_ID"].apply(lambda x: str(x).startswith('217'))]
player_game_adv_2017.to_csv(r'C:\Users\ACER\aiwork\final\train\player_game_adv_2017.csv', index = False)

player_game_adv_2018 = player_game[player_game["GAME_ID"].apply(lambda x: str(x).startswith('218'))]
player_game_adv_2018.to_csv(r'C:\Users\ACER\aiwork\final\train\player_game_adv_2018.csv', index = False)
```
