# 計算每場比賽之前對上同樣對手的對戰紀錄之平均數據
例如: <br>
&ensp; &ensp; 預測A vs. B 第十場: 
<br>
&ensp; &ensp; 用A vs. B的前九場比賽紀錄之平均數據來預測第十場的分數是多少

利用groupby找出一對一的對戰數據

```python
matchgroup = train.groupby(["Team_ID", "Match"])
```

```python
teamId = train["Team_ID"].unique()
```

取得某n場比賽的前n-1場比賽數據之平均

```python
def getMeanPerformance(teamId, matchId):
    match_before_game = pd.DataFrame()
#     按照Game_ID排序比賽(越早的比賽id越小)
    record = matchgroup.get_group((teamId, matchId)).sort_values(by=["Game_ID"])
    y = []
    for g_id in record["Game_ID"]:
        pts = record[(record["Game_ID"] == g_id) & (record["Team_ID"] == teamId)]["PTS"].values[0]
        y.append(pts)
        record_mean = record[record["Game_ID"] < g_id].groupby(["Team_ID", "Match"]).mean()
        if record_mean.empty:
            y.remove(pts)
        record_mean = record_mean.reset_index()
        match_before_game = pd.concat([match_before_game, record_mean], ignore_index=True, axis=0, sort=False)
    return match_before_game, y
```

```python
result = pd.DataFrame()
y_data = []
for team in teamId:
    for match_team in teamId:
        if team != match_team:
            df, y = getMeanPerformance(team, match_team)
            result = pd.concat([result, df], ignore_index=True, axis=0, sort=False)
            y_data += y
```
